# 6. Pemodelan Uplift

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from xgboost import XGBClassifier

# Pisah fitur dan target
X = pd.get_dummies(df.drop(columns=['conversion']), drop_first=True)
y = df['conversion']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# =======================
# 1. Logistic Regression
# =======================
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
logreg_auc = roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1])
print(f'AUC Logistic Regression: {logreg_auc:.4f}')

# =======================
# 2. Random Forest
# =======================
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_auc = roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1])
print(f'AUC Random Forest: {rf_auc:.4f}')

# =======================
# 3. XGBoost Classifier
# =======================
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)
xgb_auc = roc_auc_score(y_test, xgb.predict_proba(X_test)[:, 1])
print(f'AUC XGBoost: {xgb_auc:.4f}')
```
